```python
# Feature Engineering

# order year
df["Order Year"] = df["Order Date"].dt.year

# order month
df["Order Month"] = df["Order Date"].dt.month

# shipping days
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

# check new columns
print(df[["Order Year","Order Month","Shipping Days"]].head())
```

```python
# fix negative shipping days
df.loc[df["Shipping Days"] < 0, "Shipping Days"] = None

# remove extreme shipping days
df.loc[df["Shipping Days"] > 30, "Shipping Days"] = None

# check again
print(df["Shipping Days"].describe())
```

```python
print(df[["Order Year","Order Month","Shipping Days"]].head())
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```
